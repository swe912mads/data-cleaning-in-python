```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("imdb_cleaned.csv")
df.head()
#task1
# IMDb score distribution
plt.hist(df['imdb_score'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of IMDb Scores")
plt.xlabel("IMDb Score"); plt.ylabel("Frequency")
plt.show()

# Boxplot of Gross
sns.boxplot(x=df['gross'])
plt.title("Boxplot of Gross Income")
plt.show()

# Duration distribution
sns.histplot(df['duration'].dropna(), bins=30, kde=True)
plt.title("Distribution of Movie Duration")
plt.show()
```

```python
#task2
# Budget vs Gross
sns.scatterplot(x='budget', y='gross', data=df, alpha=0.5)
plt.title("Budget vs Gross")
plt.show()

# Genre vs IMDb Score
df['genres'] = df['genres'].astype(str).str.split('|')
df_exploded = df.explode('genres')
top_genres = df_exploded['genres'].value_counts().head(5).index
sns.boxplot(x='genres', y='imdb_score', data=df_exploded[df_exploded['genres'].isin(top_genres)])
plt.title("IMDb Score by Top 5 Genres")
plt.show()
```

```python
#task3
# Select numeric columns you want to compare
num_cols = ['imdb_score', 'gross', 'budget', 'duration', 'num_voted_users']

# Calculate correlation
corr = df[num_cols].corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()



```

result: 
- IMDb scores cluster between 67.
- Budget and Gross show positive correlation, but with high variance.
- Top genres (Drama, Comedy, Action, Thriller, Adventure) show slightly different IMDb score distributions.
- Duration tends to peak around 90120 minutes.
- Number of votes strongly correlates with IMDb score and gross.
